## Birdwatching

Here we are looking at artportalen.csv which contain observations of birds in sweden by both scientists and birdwatchers during 2022-01-01 to 2022-06-08.

```python
url2 = "https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/artportalen.csv"
birds = pd.read_csv(url2)
birds
```

We want to find the most prevalent species so we look att the number of observations of each species. to do that we first make a dataframe of just the species names column,

```python
bird_species = birds[["Artnamn"]]
bird_species
```

and then we use groupby with size to get a column of how many rows we have with that species.

```python
a = bird_species.groupby("Artnamn").size().to_frame("size")
a.sort_values("size", ascending=False)
```

And this gives us that koltrast, talgoxe and blåmes are the most prevalent species. 

We want to look at how these observations are distributed over the year so we make a dataframe of all the observations of these birds with the date they were observed and pick out the month from the date in its own column.

```python
top3birds = birds.loc[(birds["Artnamn"] == "Koltrast")|(birds["Artnamn"] == "Talgoxe")|(birds["Artnamn"] == "Blåmes")]
top3birds_cleaned = top3birds[["Artnamn","Startdatum"]]
b = pd.to_datetime(top3birds_cleaned["Startdatum"]).to_frame("Datum")
c = b["Datum"].dt.month.to_frame("Month")
d = top3birds_cleaned.join(c)
d
```

Thus we vcan groupby count the observations per species and month and make a bar plot with this.

```python
e = d.groupby(["Artnamn","Month"], as_index=False).count()
e
```

```python
sns.barplot(data=e, x="Month", y="Startdatum", hue="Artnamn")
```

Since we only hAve observations for june 1-8 month 6 is a lot smaller than the other bars, but if we just look at months 1-5 we see that blåmes and talgoxe follow eachother close, slowly declining into the year while koltrast shoots up in may.

```python
f = a.loc[a["size"] == 1]
f
```

Now we are going to look at stroke-data.csv to see if we can see anything interesting about it.

```python
url3 = "https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/stroke-data.csv"
strokes = pd.read_csv(url3)
strokes
```

We first look at a sns pairplot of the variables with more variety like age, glucose and bmi. We also sort the dataframe by strokes so those points are in front in the plots for visibility.

```python
labels = ["age","avg_glucose_level","bmi","stroke"]
sns.pairplot(strokes[labels].sort_values("stroke"), hue="stroke")
```

Looking at the plots for bmi and glucose I cant see any strong correlation but the plots with age looks to be leaning more torwards high ages for strokes so we look a bit closer at that.

We count how many individuals we have for each age and how many of those who have had a stroke and divide the strokes with the whole to get the proportion for strokes based on age, then put this in a boxplot to look at it.

```python
all_ages = strokes.groupby("age").count()
stroke_only = strokes.loc[strokes["stroke"] == 1]
strokes_only_count = stroke_only.groupby("age").count()
percent_strokes = strokes_only_count[["stroke"]]/all_ages[["stroke"]]
percent_strokes["age"] = percent_strokes.index
plt.figure(figsize=(15,5))
sns.barplot(data=percent_strokes, x="age", y="stroke")
plt.xticks(rotation=90)
plt.tight_layout()
```

From the boxplot we can see that generally the proportion of people who've had strokes goes up with age in this dataset but we have one curious bar at age 1.32 so we look at our data on 1.32 year olds,

```python
small_stroke = strokes.loc[strokes["age"] == 1.32]
small_stroke
```

and we see that we have a really small sample size(8 data points) with one of them having had a stroke making the proportion 1/8. So the high spike is more because we dont have a lot of data for this age rather than it being really likely to get a stroke at one year and four months.

Going on we look at the data with less options, like smoking or marriage.

```python
strokes["counting"] = 1
count_hypertension = strokes.groupby(["hypertension","stroke"], as_index=False).count()
count_heart_disease = strokes.groupby(["heart_disease","stroke"], as_index=False).count()
count_ever_married = strokes.groupby(["ever_married","stroke"], as_index=False).count()
count_work_type = strokes.groupby(["work_type","stroke"], as_index=False).count()
count_Residence_type = strokes.groupby(["Residence_type","stroke"], as_index=False).count()
count_smoking_status = strokes.groupby(["smoking_status","stroke"], as_index=False).count()
```

```python
plt.figure(figsize=(15,15))
plt.subplot(3,2,1)
sns.barplot(data=count_hypertension, x="hypertension" , y="counting" , hue="stroke")
plt.subplot(3,2,2)
sns.barplot(data=count_heart_disease, x="heart_disease" , y="counting" , hue="stroke")
plt.subplot(3,2,3)
sns.barplot(data=count_ever_married, x="ever_married" , y="counting" , hue="stroke")
plt.subplot(3,2,4)
sns.barplot(data=count_work_type, x="work_type" , y="counting" , hue="stroke")
plt.subplot(3,2,5)
sns.barplot(data=count_Residence_type, x="Residence_type" , y="counting" , hue="stroke")
plt.subplot(3,2,6)
sns.barplot(data=count_smoking_status, x="smoking_status" , y="counting" , hue="stroke")
```

Looking at these barplots we see that hypertension and heartdisease have a small amount of people in their 1 bars meaning that it could give the same kind of skev that or age plot had for the low ages above so they could give bad conclusions.
Residence type we can see have really alike bars for both types so we can't conclude that they matter for strokes.
Ever married does not seem significantly different in proportions but worktype and smoking status does so let's look a bit closer on them.

```python
newer_worked = {"work_type":"Newer_worked","stroke":1,"counting":0}
count_work_type_added = pd.concat([count_work_type, pd.DataFrame([newer_worked])], ignore_index=True)
bees = count_work_type_added.sort_values("work_type").reset_index(drop=True)


x_work_type = ["Govt_job","Never_worked","Private","Self-employed","children"]
y_work_type = []
for i in range(10):
    if i % 2:
        s = int(bees["counting"].iloc[i])/(int(bees["counting"].iloc[i]) + int(bees["counting"].iloc[i-1]))
        y_work_type.append(s)
        
work_type = pd.DataFrame(data=x_work_type)
work_type["proportion stroke"] = y_work_type


        
x_smoking_status = ["Unknown","formerly smoked","never smoked","smokes"]
y_smoking_status = []
for i in range(8):
    if i % 2 == 0:
        t =count_smoking_status["counting"].iloc[i+1]/(count_smoking_status["counting"].iloc[i] + count_smoking_status["counting"].iloc[i+1])
        y_smoking_status.append(t)
        
smoking_status = pd.DataFrame(data=x_smoking_status)
smoking_status["proportion stroke"] = y_smoking_status
        
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.barplot(data=work_type, x=0, y="proportion stroke").set(xlabel=None)
plt.subplot(1,2,2)
sns.barplot(data=smoking_status, x=0, y="proportion stroke").set(xlabel=None)
```

These barplots show the proportion of strokes in these categories. Newer worked and children tend to be young so their low proportions could be that. Selfemployed usually means more stress which could be a contributing factor and the same could be said for former smokers having more stress from still having the addiction but not satisfying it. 

So the possible variables that may contribute to strokes we have found so far is age, work and former smokers, two that could be tied to stress.

## Cleaning Data

Here we clean up cellphones_total.csv so that all values are numbers and not strings.

```python
url4 = "https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/cell_phones_total.csv"
cellphones = pd.read_csv(url4)
cellphones
```

So we define a function that takes a value as input, if it already is a float or int it returns that number, if it is a string it looks for K M and B in the string, removes the letter and multiply the remaining number with the apropriate power of ten. then we make a new dataset with the country codes and run trough the old dataset column by column applying our function and joining our new columns to the new dataset.

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

cellphones1 = cellphones[["iso-3"]]
cellphones_temp = cellphones["1960"].apply(value_to_float)
cellphones1["1960"] = cellphones_temp


for i in range(1965,2020):
    cellphones_temp = cellphones[str(i)].apply(value_to_float)
    cellphones1 = cellphones1.join(cellphones_temp)


cellphones1
```

And this gives us a dataframe with numbers for all the relevant values
